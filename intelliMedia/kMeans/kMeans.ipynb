```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#data(iris) import
data = pd.read_csv('iris.csv')
x = data.iloc[:,0:4].values
```

```python
#クラスタ数1~10でk-meansを実行
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#plot
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within cluster sum of squares')
plt.show()
plt.savefig('TheElbowMethod.png')
```

```python
#有効と思われるk=3で実行
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)

plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s=100, c='green', label='Iris-versicolour')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s=100, c='red', label='Iris-setosa')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s=100, c='blue', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='yellow', label='Centroids')
plt.legend()
plt.savefig('cluster3.png')
```
