```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#標準正規分布（平均0, 標準偏差1を64個作成）
score0 = np.random.normal(0, 1, 64)
score1 = np.random.normal(0, 1, 64)
score2 = np.random.normal(0, 1, 64)
score3 = np.random.normal(0, 1, 64)
score4 = np.random.normal(0, 1, 64)

print(score0)
```

```python
plt.hist(score0, bins=10)
plt.xlim(-5,5)
```

```python
for i in range(64):
    if i==0:
        continue
    else:
        score0[i] += score0[i-1]
        score1[i] += score1[i-1]
        score2[i] += score2[i-1]
        score3[i] += score3[i-1]
        score4[i] += score4[i-1]
print(score0)
```

```python
sns.set()
plt.plot(score0, label='score0')
plt.plot(score1, label='score1')
plt.plot(score2, label='score2')
plt.plot(score3, label='score3')
plt.plot(score4, label='score4')

#グラフの軸
plt.xlabel('TIME')
plt.ylabel('Score')
#グラフの凡例
plt.legend()

plt.show()
```

```python
print(score0[0], '\t', score0[63], '\t差分：', score0[63]-score0[0])
print(score1[0], '\t', score1[63], '\t差分：', score1[63]-score1[0])
print(score2[0], '\t', score2[63], '\t差分：', score2[63]-score2[0])
print(score3[0], '\t', score3[63], '\t差分：', score3[63]-score3[0])
print(score4[0], '\t', score4[63], '\t差分：', score4[63]-score4[0])
```
